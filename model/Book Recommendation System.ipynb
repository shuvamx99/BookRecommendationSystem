# Collaborative Filtering Based Recommendation System

```python
ratings_with_name
```

```python
# Trying to get users which have rated a lot of books. Users who have rated 200 books
```

```python
ratings_with_name.groupby('User-ID').count()
```

```python
ratings_with_name.groupby('User-ID').count()['Book-Rating']
```

```python
# Users who have rated atleast 200 books
```

```python
x = ratings_with_name.groupby('User-ID').count()['Book-Rating']>200
x[x] #boolean indexing
```

```python
# Extract the ids of those users
```

```python
veteran_reviewers = x[x].index
```

```python
veteran_reviewers
```

```python
# Filtering on basis of the veteran users
```

```python
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(veteran_reviewers)]
```

```python
# We will further filter the data on book-rating
```

```python
filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
```

```python
y=filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
y[y]
```

```python
famous_books = y[y].index
```

```python
famous_books
```

```python
filtered_rating['Book-Title'].isin(famous_books)
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_ratings.drop_duplicates()
```

```python
pt = final_ratings.pivot_table(index = 'Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.shape
```

```python
pt
```

```python
# There are 706 Books and 810 Users
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
# We can consider each book to be a vector in a 810-D space
```

```python
# We find similar books we calculate the euclidean distance of these vectors
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(pt) #Euclidean distance of those 706 books
```

```python
cosine_similarity(pt).shape
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores[0]
```

```python
np.where(pt.index=='Year of Wonders')[0][0]
```

```python
index = np.where(pt.index=='Year of Wonders')[0][0]
similarity_scores[index]
```

```python
list(enumerate(similarity_scores[index]))
```

```python
sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:6]
```

```python
# Finding books with most similarity with chosen book, picking top 5^
```

```python
def recommend(book_name):
    index = np.where(pt.index == book_name)[0][0]
    similar_books = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:9]
    data=[]
    for i in similar_books:
        item=[]
        temp_df = (books[books['Book-Title'] == pt.index[i[0]]])
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
    return data
    
```

```python
recommend('Harry Potter and the Prisoner of Azkaban (Book 3)')
```

```python
recommend('The Two Towers (The Lord of the Rings, Part 2)')
```

```python
recommend('Life of Pi')
```
